##We listed commonly used text styles and the corresponding markdown syntax in the format text section.

- _**Headings**_ are created by adding a pound sign (#) at the beginning.
- _**Bold**_ is achieved by adding two stars (**) at both ends. (**This is bold**)
- _**Italic**_ is achieved by adding one star (*) at both ends. *This is italicized*
- _**Strikethrough**_ is achieved by adding two tildes (~~) at both ends. ~~This is a strikethrough~~
- _**Blockquote**_ is achieved by adding one greater than sign (>) at the beginning. > This is a blockquote
- _**Inline code**_ is achieved by adding one backtick (`) at both ends. `This is an inline code`
- A _**code block**_ is achieved by adding a triple backtick (```) at both ends.
